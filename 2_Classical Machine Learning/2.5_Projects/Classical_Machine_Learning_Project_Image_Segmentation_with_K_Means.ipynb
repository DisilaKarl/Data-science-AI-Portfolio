```python
from PIL import Image
import numpy as np
```

```python
# Loading image and converting to RGB
img = Image.open('Apple.jpg')

# Convert image to numpy array
img_array = np.array(img)
```

```python
# Array feature

print(img_array.shape)
```

**2. Implementing the K-means Algorithm for Image Segmentation**

```python
from sklearn.cluster import KMeans
```

```python
# Reshape the image into a 2D array where each row is a pixel
pixels = img_array.reshape(-1, 3) # Changed 'image_array' to 'img_array'

# Number of clusters (k)
k = 5

# Apply the K-means algorithm
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(pixels)

# Get the centroids and labels for the clusters
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Convert centroids to integers
centroids = centroids.round(0).astype(int)

# Create the segmented image by assigning each pixel the color of its centroid
segmented_image = centroids[labels].reshape(img_array.shape) # Changed 'image_array' to 'img_array'
```

```python
# Display the segmented image
import matplotlib.pyplot as plt
plt.imshow(segmented_image)
plt.show()
```

**3. Saving the Segmented Image**

```python
# Convert the segmented image back into a PIL image
# Convert the data type to uint8 to be compatible with PIL
segmented_image_pil = Image.fromarray(segmented_image.astype(np.uint8))

# Save the segmented image
segmented_image_pil.save('segmented_image.jpg')  # Save as 'segmented_image.jpg'
```

**4. Limitations and potential improvements of the implemented K-Means algorithm for image segmentation.**

# Limitation

1. Choosing the Right Number of Clusters (k):

K-means requires us to choose how many clusters (groups) we want before running the algorithm. This is tricky because if we choose the wrong number of clusters, the segmentation might not look good. Its like trying to divide a group of people into teams, but you dont know how many teams you need!

2 . Struggles with Noise and Outliers:

Noise means random, irrelevant data (like a few scattered bright pixels that dont belong). K-means has trouble with this because it tries to fit everything into clusters, even if some data points (like noisy pixels) don't belong. These outliers can mess up the centroids and make the segmentation look worse.

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns


# Option pour un affichage plus esth√©tique avec Seaborn

sns.set(style="whitegrid")
```

```python
pokemon_df = pd.read_csv('/content/Pokemon.csv')
```

```python
print(pokemon_df.head())
```

```python
#Scatter Plot: Attack vs Defense
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pokemon_df, x='Attack', y='Defense')
plt.title('Attack vs Defense Stats of Pokemon')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.savefig('attack_vs_defense.png')
plt.show()
```

```python
# 2. Histogram: Distribution of Speed Stat
plt.figure(figsize=(8, 6))
sns.histplot(pokemon_df['Speed'], kde=True)
plt.title('Distribution of Pokemon Speed')
plt.xlabel('Speed')
plt.ylabel('Frequency')
plt.savefig('speed_distribution.png')
plt.show()
```

```python
# 3. Line Plot: Average Attack and Defense Over Time (Generation)
plt.figure(figsize=(8, 6))
avg_stats = pokemon_df.groupby('Generation')[['Attack', 'Defense']].mean().reset_index()
sns.lineplot(data=avg_stats, x='Generation', y='Attack', label='Attack', marker='o')
sns.lineplot(data=avg_stats, x='Generation', y='Defense', label='Defense', marker='o')
plt.title('Average Attack and Defense Stats Over Generations')
plt.xlabel('Generation')
plt.ylabel('Average Stat')
plt.legend()
plt.savefig('average_attack_defense.png')
plt.show()
```

```python
# 4. Bar Plot: Legendary vs Non-Legendary Pokemon
plt.figure(figsize=(8, 6))
sns.countplot(data=pokemon_df, x='Legendary')
plt.title('Number of Legendary and Non-Legendary Pokemon')
plt.xlabel('Legendary')
plt.ylabel('Count')
plt.savefig('legendary_vs_nonlegendary.png')
plt.show()
```

```python
# 5. Heatmap: Correlation Matrix
plt.figure(figsize=(10, 8))
corr_matrix = pokemon_df[['Attack', 'Defense', 'Speed', 'HP', 'Sp. Atk', 'Sp. Def']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Pokemon Stats')
plt.savefig('correlation_heatmap.png')
plt.show()
```

```python
# Bonus 1: Pair Plot of All Numeric Variables
sns.pairplot(pokemon_df[['Attack', 'Defense', 'Speed', 'HP', 'Sp. Atk', 'Sp. Def']])
plt.savefig('pair_plot.png')
plt.show()
```

```python
# Bonus 2: Violin Plot of Pokemon Types and Attack Stats
plt.figure(figsize=(12, 6))
sns.violinplot(data=pokemon_df, x='Type 1', y='Attack')
plt.title('Distribution of Attack Stats by Pokemon Type')
plt.xlabel('Pokemon Type')
plt.ylabel('Attack Stat')
plt.xticks(rotation=45)
plt.savefig('attack_by_type.png')
plt.show()
```
